# Número óptimo de Clusters

- Seleccione una (una sola) medida para optener el número optimo de grupos que deberia su dataset (grafique)

```python
datos_SP = datos[['Error1', 'Error2', 'Error3', 'Accepted', 'Parcial1', 'Parcial2',
       'Final', 'Tiempo','Entregas']]
datos_SP
```

```python
# Calculating the inertia and silhouette_score¶
inertia = []
sil = []
# changing the number of clusters 
for k1 in range(2,11):
    modelo1 = KMeans(n_clusters=k1,random_state = 42)
    modelo1.fit(datos_SP)
    y_pred_gano = modelo1.predict(datos_SP)
    
    inertia.append((k1, modelo1.inertia_))
    sil.append((k1, silhouette_score(datos_SP, y_pred_gano)))
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
# Plotting Elbow Curve
x_iner = [x[0] for x in inertia]
y_iner  = [x[1] for x in inertia]
ax[0].plot(x_iner, y_iner)
ax[0].set_xlabel('Número of Clusters')
ax[0].set_ylabel('Intertia')
ax[0].set_title('Curva Elbow')
# Plotting Silhouetter Score
x_sil = [x[0] for x in sil]
y_sil  = [x[1] for x in sil]
ax[1].plot(x_sil, y_sil)
ax[1].set_xlabel('Número de Clusters')
ax[1].set_ylabel('Silhouetter Score')
ax[1].set_title('Curva Silhouetter')
```

```python
#Creamos el modelo 
from sklearn.cluster import KMeans

#seleecion de centroides
modelo = KMeans(n_clusters = 2, random_state = 42)
```
