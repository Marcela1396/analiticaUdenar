## Comprensión  y preparación de los Datos

Carga, exploración y transformación de los datos

El dataset se compone de 657 registros de 5 cursos de A&G que se dictaron en el año 2019 cohorte II, los cuales pertenecen a cinco programas de la Facultad de Ingeniería de la Universidad del Valle


```python
#Carga de Librerias
import pandas as pd
import numpy as np
```

```python
#Carga los datos provenientes de un csv
datos = pd.read_csv("estudiantes.csv", sep=';')
```

```python
datos
```

```python
datos.describe()
```

```python
#Se revisa la cantidad de nulos por cada columna
datos.isnull().sum()
```

```python
# Se elimina las columnas que no son relevantes para el objeto de estudio
datos = datos.drop(['No.', 'Programa', 'T. Mat.', 'Curso','Periodo','semestre','Nota1','Envios','pasoparcial','Grupo'], axis=1)
datos
```

```python
#Se revisa la cantidad de nulos por cada columna que se va a considerar dentro del estudio
datos.isnull().sum()
```

```python
#Cambiar el nombre a las columnas cuyo nombre es extenso
datos.rename(columns={'Error1 (Wrong Answer)':'Error1',
                      'Error2 (Runtime bad)':'Error2',
                      'Error3 (exeption error)':'Error3',
                      'Accepted':'Aceptado',
                      'Gano curso?':'GanaCurso'
                     },
               inplace=True)
datos.columns
```

```python
datos
```

```python
datos['Error1'] = datos['Error1'].replace([np.nan], 0)
datos['Error2'] = datos['Error2'].replace([np.nan], 0)
datos['Error3'] = datos['Error3'].replace([np.nan], 0)
```

```python
datos
```

```python
datos.isnull().sum()
```

```python
# Se crea una nueva columna llamada Entregas la cual se obtiene con la sumatoria de todos las
# columnas : Error1, Error2, Error3, Accepted
datos['Entregas'] = datos['Error1'] + datos['Error2'] + datos['Error3'] + datos['Accepted']
```

```python
datos
```

```python
datos["Final"] = [x.replace(",", ".") for x in datos["Final"]]
datos["Final"] = pd.to_numeric(datos["Final"], downcast="float")

datos["Parcial1"] = [x.replace(",", ".") for x in datos["Parcial1"]]
datos["Parcial1"] = pd.to_numeric(datos["Parcial1"], downcast="float")

datos["Parcial2"] = [x.replace(",", ".") for x in datos["Parcial2"]]
datos["Parcial2"] = pd.to_numeric(datos["Parcial2"], downcast="float")

datos
```

```python
datos.describe()
```

```python
data= datos.describe()
data.to_csv (r'estadisticas.csv', index = False, header=True)
```
